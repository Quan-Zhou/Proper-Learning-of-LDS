```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
colormap=('#4292c6','#BDB2FF', '#FF99C8' ,'#EEB422')
labels=['least squares auot','subspace auto','ssarx','our method']
```

# Higher dimensions

```python
mean=pd.read_csv("data/ncpop300_higherdim_mean_plot.csv",header=None,index_col=None)
std=pd.read_csv("data/ncpop300_higherdim_std_plot.csv",header=None,index_col=None)
met_range=[*range(4)]
nx_range=np.arange(2,5)
```

```python
fig, ax = plt.subplots()
width=0.2
for m in met_range: 
    x=nx_range+m*width
    y=mean.loc[m,:]
    y_error=mean.loc[m,:]-std.loc[m,0:2]
    #ax.bar(x,2, color=colormap[m], width=width,bottom=y.tolist())
    ax.plot(x,y.tolist(), color=colormap[m],label=labels[m])
    ax.errorbar(x, y,yerr = y_error, fmt ='o', color=colormap[m]) #,align='edge'

#plt.legend(bbox_to_anchor=(1.02,1.25),fontsize=14,frameon=False,ncol=2)
plt.xticks(ticks=[2.3,3.3,4.3],labels=[2,3,4])
plt.xlabel('dimensions of system matrices '+r'$n$',fontsize=14)
plt.ylabel('nrmse(%)\n of higher dimensional systems',fontsize=14)
plt.savefig('prettyplot_higherdim.pdf', bbox_inches='tight') 
```

# Higher differencial order

```python
mean=pd.read_csv("data/ncpop300_highersd_mean_plot.csv",header=None,index_col=None)
std=pd.read_csv("data/ncpop300_highersd_std_plot.csv",header=None,index_col=None)
met_range=[*range(4)]
para_range=np.arange(0.1,1,0.1)
```

```python
fig, ax = plt.subplots()
width=0.2
for m in met_range: 
    x=para_range#+m*width
    y=mean.loc[m,:]
    yup=std.loc[m,0:8]
    ydown=std.loc[m,9:17]
    ax.plot(x,y.tolist(), color=colormap[m],label=labels[m])
    ax.plot(x,yup.tolist(), color=colormap[m],linestyle='--',linewidth=0.8)
    ax.plot(x,ydown.tolist(), color=colormap[m],linestyle='--',linewidth=0.8)
    #ax.errorbar(x, y,yerr = y_error, fmt ='o', color=colormap[m]) #,align='edge'

#plt.legend(bbox_to_anchor=(1.02,1.25),fontsize=14,frameon=False,ncol=2)
plt.xticks(ticks=para_range)
plt.xlim([0.1,0.9])
plt.yticks(ticks=[-100,-50,0,50,100])
plt.xlabel('noise std '+r'$W=V$',fontsize=14)
plt.ylabel('nrmse(%)\n of higher order systems',fontsize=14)
plt.savefig('prettyplot_highersd.pdf', bbox_inches='tight') 
```
