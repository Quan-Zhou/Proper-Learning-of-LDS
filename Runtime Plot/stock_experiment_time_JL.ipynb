```julia
using TSSOS
using DynamicPolynomials
using CSV
using Random
using Plots
using Statistics
using DataFrames
using DelimitedFiles
```

```julia
# read observations
Y= readdlm( "stock-market.txt" );
```

```julia
function ncpop(T)

# set nc operators
@ncpolyvar G Fdash m[1:T+1] q[1:T] p[1:T] f[1:T]; # all varaibles are assumed to be nc
var=vcat(p,f,G,Fdash,m,q);

# constraints
ine1 = [f[i] - p[i] - Fdash*m[i+1] for i in 1:T];
ine2 = [- f[i] + p[i] + Fdash*m[i+1] for i in 1:T];
ine3 = [m[i+1] - q[i] - G*m[i] for i in 1:T];
ine4 = [- m[i+1] + q[i] + G*m[i] for i in 1:T];

#objective
obj=-sum((Y[i]-f[i])^2 for i in 1:T)+0.1*sum(p[i]^2 for i in 1:T) ;

# pop
pop=vcat(obj,ine1,ine2,ine3,ine4);

# solve model 
opt,sol,data=blockcpop_first(pop,var,1,method="chordal");

end
```

```julia
meanR=Float64[]
stdR=Float64[]

for t in 5:30
    Re=Float64[]
    for r in 1:3
        timeR=@elapsed ncpop(t);
        push!(Re,timeR)
    end
    
    push!(meanR,mean(copy(Re)))
    push!(stdR,std(copy(Re)))
    
end
```
